```python
from pycaret.nlp import *
```

```python
# Grabs topics from the dataset and classifies using LDA
df = get_topics(data=train, text='Text')
```

```python
ml = create_model(model='lda', multi_core=True)
```

```python
#assigns lda to dataset and preps for nlp classification
lda_model = assign_model(ml)
```

```python
lda_model.head()
```

```python
#Evaluates all important metrics
evaluate_model(ml)
```

```python
#Fits model to the dataset and target
lda_model_1 = lda_model[['Corrected_label', 'Topic_0', 'Topic_1', 'Topic_2', 'Topic_3']]
```

```python
lda_model_1
```

```python
from pycaret.classification import *
```

```python
#Sets up classification model
cla = setup(data = lda_model_1, target='Corrected_label', session_id=123, train_size=0.80, log_experiment=True, experiment_name='nlp')
```

```python
#Previews best models based on dataset metrics
best = compare_models()
```

```python
#Models for dataset regression including random forest, lightgbm, extra trees, linear discriminant analysis, and gradient boost along with accuracy scores, auc, recall, precision, F1 scores, Kappa, and MCC scores
cla_rf = create_model('rf')
cla_lgbm = create_model('lightgbm')
cla_et = create_model('et')
cla_gbc = create_model('gbc')
cla_lda = create_model('lda')
```

```python
#Tuned models with metrics
tuned_cla_rf = tune_model(cla_rf)
tuned_cla_lgbm = tune_model(cla_lgbm)
tuned_cla_et = tune_model(cla_et)
tuned_cla_gbc = tune_model(cla_gbc)
tuned_cla_lda = tune_model(cla_lda)
```

```python
#This cell and the next four cells provide accuracy metrics of the tuned models starting with random forest, then lightgbm, then extra trees, gradient boost and finally lda
evaluate_model(tuned_cla_rf)
evaluate_model(tuned_cla_lgbm)
evaluate_model(tuned_cla_et)
evaluate_model(tuned_cla_gbc)
evaluate_model(tuned_cla_lda)
```
