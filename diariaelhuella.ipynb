```
!pip3 install newspaper3k
import requests
import nltk
nltk.download('punkt')
from bs4 import BeautifulSoup
from newspaper import Article
import pandas as pd
```

```
feed = 'https://diariolahuella.opennemas.com/rss/politica'

response = requests.get(feed)

website = response.content

soup = BeautifulSoup(website, features='xml')

items = soup.find_all('item')

articles = []
for item in items:
    link = item.find('link').text
    articles.append(link)

    
```

```
from newspaper import Article
```

```
multi_pages = []
for url in articles:
    article = Article(url)
    article.download()
    article.parse()
    article.nlp()

    df = pd.DataFrame()
    titles = []
    links = []
    content_text =[]
    published_on = []

    title = article.title
    text = article.text
    keywords = article.keywords
    summary = article.summary
    date = article.publish_date
    authors = article.authors

    titles.append(title)
    links.append(url)
    content_text.append(summary)
    published_on.append(date)

        
    print('************************************************************')
    print(f"Title: {title}")
    print(f"URL: {url}")
    print(f"Authors: {authors}")
    print(f"Date: {date}")
    print(f"Keywords: {keywords}")
    print(f"Summary: {summary}")
    print(f"Text: {text}")
    print('************************************************************')

    df['title'] = titles
    df['link'] = links
    df['content_text'] = content_text
    df['published_on'] = published_on
    multi_pages.append(df)
 
df_new = pd.concat(multi_pages).reset_index(drop =True)
df_new.head()




```
