```python
#Models for dataset regression including random forest, lightgbm, extra trees, linear discriminant analysis, and gradient boost along with accuracy scores, auc, recall, precision, F1 scores, Kappa, and MCC scores
cla_rf = create_model('rf')
cla_lgbm = create_model('lightgbm')
cla_et = create_model('et')
cla_gbc = create_model('gbc')
cla_lda = create_model('lda')
```

```python
#Tuned models with metrics
tuned_cla_rf = tune_model(cla_rf)
tuned_cla_lgbm = tune_model(cla_lgbm)
tuned_cla_et = tune_model(cla_et)
tuned_cla_gbc = tune_model(cla_gbc)
tuned_cla_lda = tune_model(cla_lda)
```

```python
test = pd.read_parquet("/content/df1dafd7a13176813587af8ca6367f723827be4a93522907e443925340f4e445")
```

```python
test.head()
```

```python
test_df = get_topics(data=test, text='Text')
```

```python
stacker = stack_models([tuned_cla_rf, tuned_cla_lda, tuned_cla_lgbm, tuned_cla_et, tuned_cla_gbc])
```

```python
test_df.head()
```

```python
test_df_model = test_df[['Prediction', 'Topic_0', 'Topic_1', 'Topic_2', 'Topic_3']]
```

```python
#Provides predictions based on a tuned stacked model
predictions = predict_model(stacker, data=test_df_model, raw_score=True)
predictions.head()
```
