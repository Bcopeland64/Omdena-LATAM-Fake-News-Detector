```python
from pycaret.nlp import *
```

```python
#Grabs topics from the dataset and classifies using LDA
df = get_topics(data=fake, text='Content')
```

```python
ml = create_model(model='lda', multi_core=True)
```

```python
#assigns lda to dataset and preps for nlp classification
lda_model = assign_model(ml)
```

```python
lda_model.head()
```

```python
#Evaluates all important metrics
evaluate_model(ml)
```

```python
#Fits model to the dataset and target
lda_model_1 = lda_model[['Prediction', 'Topic_0', 'Topic_1', 'Topic_2', 'Topic_3']]
```

```python
lda_model_1
```

```python
from pycaret.classification import *
```

```python
#Sets up classification model
cla = setup(data = lda_model_1, target='Prediction', session_id=123, train_size=0.80)
```

```python
#Previews best models based on dataset metrics
best = compare_models()
```

```python
#Models for dataset regression including linear regression, lda, naive bayes, and support vector machines along with accuracy scores, auc, recall, precision, F1 scores, Kappa, and MCC scores
cla_lr = create_model('lr')
cla_lda = create_model('lda')
cla_nb = create_model('nb')
cla_svm = create_model('svm')
```

```python
#Tuned models with metrics
tuned_cla_lr = tune_model(cla_lr)
tuned_cla_lda = tune_model(cla_lda)
tuned_cla_nb = tune_model(cla_nb)
tuned_cla_svm = tune_model(cla_svm)
```

```python
#This cell and the next four cells provide accuracy metrics of the tuned models starting with linear regression, then lda, then naive bayes, and finally support vector machines
evaluate_model(tuned_cla_lr)
```
